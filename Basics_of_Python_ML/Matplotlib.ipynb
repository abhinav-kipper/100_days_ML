#Customising Graphs

```python
#plot function x**3
#color marker linewidth label for axis
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
#plt.plot(x,y,color='black',marker='p')
plt.plot(x,y,color='black',linewidth=5,label='x^3')
plt.plot(x,y2,color='red',linewidth=5,label='x^2')

plt.ylabel("y")
plt.xlabel("x")

plt.legend() # to view labels
plt.title("Matplotlib demo")

plt.grid()

plt.text(2,80,"HERE",fontsize=12)

plt.axis([0,10,0,200]) # 0 10 for x 0 200 for y
plt.show()
```

#PIE GRAPH

```python
labels=["A","B","C","D"]
sizes=[3,4,6,2]
colors=["blue","purple","red","pink"]
plt.pie(sizes,colors=colors,labels=labels)
plt.axis("equal")
plt.title("SPLIT among CLASSES")
plt.show()
```
