```python
df=iris.copy()
df.columns=['sepall','sepalw','petall','petalw','flower_type']
df.tail(4)
```

```python
#ADD a column
df["diff_petall_petalw"]=df["petall"]-df["petalw"]
df.tail()
```

```python
#Handling NANS
#option 1 Drop Nan row
#option 2 Fill Nan
```

```python
#adding nans
import numpy as np
df.iloc[2:4,1:3]=np.nan
df.head()
```

```python
df.describe()
```

```python
#option 1 
df.dropna(inplace=True)
#dropped nans and resetted index
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
#option2 filling
#Fill these nans with the  mean of that column or Most occuring value or mean of that particular flower_type
df.head()
```

```python
df.sepalw.fillna(df.sepalw.mean(),inplace=True)
df.head()
```

```python
a=df[df.flower_type=='Iris-setosa'] #mean of only those with this flower type
df.petall.fillna(a.petall.mean(),inplace=True)
df.head()
```

```python
#Handling strings in data
df['Gender']='Female'
df.iloc[0:10,5]='Male'
df.head(12)

def f(s):
    if s=='Male':
        return 0
    return 1
df['sex']=df.Gender.apply(f) #takes function and applies it to all entries
df.head(12)
```
