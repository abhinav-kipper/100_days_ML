```python
import pandas as pd
df=pd.read_csv('powerplant_train.csv')

test=pd.read_csv('powerplant_test.csv')
#directly putted coclumns  in csv file to save first datapoint
#test.columns=['T','V','AP','RH']
test
df.columns=['T','V','AP','RH','EP']
df['T_T']=df['T']**2
df['V_V']=df['V']**2
df['AP_AP']=df['AP']**2
df['RH_RH']=df['RH']**2
df['T_V']=df['T']*df['V']
df['T_AP']=df['T']*df['AP']
df['T_RH']=df['T']*df['RH']
df['RH_AP']=df['RH']*df['AP']
df['AP_V']=df['AP']*df['V']
df['RH_V']=df['RH']*df['V']
target=df.EP
data=df.iloc[:,:-1]
test.shape
data.shape
target.shape
df
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
clf.fit(data,target)
```

```python
import numpy as np


test['T_T']=test['T']**2
test['V_V']=test['V']**2
test['AP_AP']=test['AP']**2
test['RH_RH']=test['RH']**2
test['T_V']=test['T']*test['V']
test['T_AP']=test['T']*test['AP']
test['T_RH']=test['T']*test['RH']
test['RH_AP']=test['RH']*test['AP']
test['AP_V']=test['AP']*test['V']
test['RH_V']=test['RH']*test['V']
ytest=clf.predict(test)
type(ytest)
ytest.shape
np.savetxt('GDpowerplantoutput.csv',ytest,delimiter=',')
ytest
```
